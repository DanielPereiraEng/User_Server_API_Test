```python
import tkinter as tk
from tkinter import messagebox
import requests

class LocationApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Localização do Usuário")

        # Labels e campos de entrada para latitude e longitude
        self.label = tk.Label(root, text="Por favor, informe sua localização:")
        self.label.pack()

        self.latitude_label = tk.Label(root, text="Latitude:")
        self.latitude_label.pack()
        self.latitude_entry = tk.Entry(root)
        self.latitude_entry.pack()

        self.longitude_label = tk.Label(root, text="Longitude:")
        self.longitude_label.pack()
        self.longitude_entry = tk.Entry(root)
        self.longitude_entry.pack()

        # Botão para enviar a localização
        self.submit_button = tk.Button(root, text="Enviar", command=self.submit_location)
        self.submit_button.pack()

    def submit_location(self):
        # Obter latitude e longitude inseridas pelo usuário
        latitude = self.latitude_entry.get()
        longitude = self.longitude_entry.get()
        if latitude and longitude:
            try:
                # Enviar solicitação à API com latitude e longitude
                response = requests.get(f"http://127.0.0.1:5000/locations?latitude={latitude}&longitude={longitude}")
                data = response.json()
                if response.status_code == 200:
                    # Se a solicitação for bem-sucedida, fechar a janela atual e abrir a próxima
                    self.root.destroy()
                    app = InformationApp(data)
                else:
                    # Se a API retornar um erro, exibir uma mensagem de erro
                    messagebox.showerror("Erro", data["error"])
            except requests.exceptions.RequestException as e:
                # Se houver um erro de conexão, exibir uma mensagem de erro
                messagebox.showerror("Erro de Conexão", f"Erro ao conectar à API: {e}")
        else:
            # Se algum campo estiver vazio, exibir uma mensagem de erro
            messagebox.showerror("Erro", "Por favor, preencha todos os campos.")

class InformationApp:
    def __init__(self, data):
        self.data = data

        # Criar uma nova janela para exibir as informações úteis
        self.root = tk.Tk()
        self.root.title("Informações Úteis")
        
        # Definir o tamanho da janela
        self.root.geometry("400x300")

        # Rótulo para exibir as informações úteis
        self.label = tk.Label(self.root, text="Informações Úteis:")
        self.label.pack()

        # Exibir as informações recebidas da API
        self.display_data()

        # Botão para sair do programa
        self.quit_button = tk.Button(self.root, text="Sair", command=self.root.destroy)
        self.quit_button.pack()

    def display_data(self):
        # Exibir as informações em rótulos individuais
        for key, value in self.data.items():
            label = tk.Label(self.root, text=f"{key}: {value}")
            label.pack()

if __name__ == "__main__":
    root = tk.Tk()
    # Definir o tamanho padrão da janela
    root.geometry("400x300")
    app = LocationApp(root)
    root.mainloop()
```
